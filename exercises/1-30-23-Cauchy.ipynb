```python
import matplotlib.pyplot as plt
import pandas as pd
import h5py
import numpy as np
import scipy.stats as stats
import statsmodels.api as api
```

```python
## Generate random numbers from normal distribution

from scipy.stats import norm, uniform, cauchy
import numpy as np
normal = norm.rvs(size=100)
uni = uniform.rvs(size=100)
cauch = cauchy.rvs(size=100)
```

```python
api.qqplot(normal, norm, color='blue',label='Normal')
api.qqplot(uni, uniform, color='green',label='Uniform')
api.qqplot(cauch, cauchy, color='orange',label='Cauchy')
```
